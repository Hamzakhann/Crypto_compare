```python
import pandas as pd
import datetime as dt
import requests
import matplotlib.pyplot as plt

getURL = requests.get('https://min-api.cryptocompare.com/data/histominute?fsym=ETH&tsym=PKR&limit=60&aggregate=1&e=CCCAGG').json()['Data']
df = pd.DataFrame(getURL)
df['timestamp'] = [dt.datetime.fromtimestamp(x) for x in df.time]
plt.plot(df['timestamp'], df['high'])
plt.show()
print(df)
```
